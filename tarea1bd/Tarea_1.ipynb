## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director (
    id_director INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER NOT NULL,
    PRIMARY KEY (id_director)
        
);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor (
    id_actor INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER NOT NULL,
    PRIMARY KEY (id_actor)
);
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER NOT NULL,
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INTEGER,
    fecha_publicacion DATE,
    PRIMARY KEY (id_pelicula),
    FOREIGN KEY (id_director) REFERENCES director(id_director)
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
    id_actor INTEGER NOT NULL ,
    id_pelicula INTEGER NOT NULL,
    PRIMARY KEY (id_actor, id_pelicula),
    FOREIGN KEY (id_actor) REFERENCES actor(id_actor),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER NOT NULL ,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio),
    FOREIGN KEY (id_director) REFERENCES director(id_director)
);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_actor INTEGER NOT NULL,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio),
    FOREIGN KEY (id_actor) REFERENCES actor(id_actor)
);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (
    id_premio INTEGER  GENERATED ALWAYS AS IDENTITY,
    id_pelicula INTEGER NOT NULL,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio),
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
);
```

**Solución:** La consulta listada selecciona los nombres de los actores que estan en películas del género 'Comedia'. Utiliza dos INNER JOIN para las tablas actor, actor_participa_en_pelicula, y pelicula, asegurando que se consideren las películas que corresponden al género especificado.

### Consulta 4
**Nombre y cantidad de premios ganados por cada género, ordenados de mayor a menor según la cantidad de premios.**


**Ejemplo:**
|Género  |Cantidad de premios|
|--------|-------------------|
|Drama   |99                 |
|Fantasía|29                 |
|...     |...                |



**Solución:** La consulta agrupa películas por género y obtiene la cantidad de premios que cada género ha ganado, usando un INNER JOIN para las tablas pelicula y premio_mejor_pelicula. Luego, ordena los resultados de mayor a menor según la cantidad de premios.

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */

SELECT pelicula.genero AS "Género", COUNT(premio_mejor_pelicula.id_premio) AS "Cantidad de premios"
FROM pelicula
INNER JOIN premio_mejor_pelicula ON pelicula.id_pelicula=premio_mejor_pelicula.id_pelicula
GROUP BY pelicula.genero
ORDER BY "Cantidad de premios" DESC;
```

**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**

Luego, agrupa los resultados por película y solo considera aquellas con dos o más actores, ordenando los resultados de mayor a menor según la cantidad de actores.

```python
%%sql
/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT pelicula.nombre AS "Nombre", COUNT(actor_participa_en_pelicula.id_actor) AS "Cantidad de actores"
FROM pelicula
INNER JOIN actor_participa_en_pelicula ON pelicula.id_pelicula=actor_participa_en_pelicula.id_pelicula
GROUP BY pelicula.id_pelicula, pelicula.nombre
HAVING COUNT(actor_participa_en_pelicula.id_actor) >= 2
ORDER BY "Cantidad de actores" DESC;
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).


**Ejemplo:**
|Nombre|Edad del actor más joven|Edad del actor más viejo|

**Solución:** Esta consulta identifica la película con la mayor diferencia de edad entre el actor más joven y el más viejo que participaron en ella. Selecciona el nombre de la película, la edad del actor más joven y la edad del actor más viejo, usando INNER JOIN para las tablas pelicula, actor_participa_en_pelicula, y actor.








```python
%%sql
SELECT pelicula.nombre AS "Nombre",
       MIN(actor.edad) AS "Edad del actor más joven",
       MAX(actor.edad) AS "Edad del actor más viejo"
FROM pelicula
INNER JOIN actor_participa_en_pelicula ON pelicula.id_pelicula = actor_participa_en_pelicula.id_pelicula
INNER JOIN actor ON actor_participa_en_pelicula.id_actor = actor.id_actor
GROUP BY pelicula.nombre
ORDER BY (MAX(actor.edad) - MIN(actor.edad)) DESC
LIMIT 1;
```

### Consulta 8
**Nombre de todos los actores que han trabajado con más de un director, la cantidad de directores con las que han trabajado y el nombre de una de las películas.**


**Ejemplo:**
|Nombre del actor|Cantidad de directores|Nombre de una película|
|----------------|----------------------|----------------------|
|Cillian Murphy  |29                    |Oppenheimer           |
|...             |...                   |...                   |


**Solución:** Esta consulta identifica actores que han trabajado con más de un director. Selecciona el nombre del actor, cuenta el número de directores con los que ha trabajado, y muestra el nombre de una de las películas en las que ha actuado.

```python
%%sql
SELECT 
    actor.nombre AS "Nombre del actor",
    COUNT(DISTINCT pelicula.id_director) AS "Cantidad de directores",
    MAX(pelicula.nombre) AS "Nombre de una película"
FROM 
    actor
JOIN 
    actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
JOIN 
    pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
GROUP BY 
    actor.id_actor
HAVING 
    COUNT(DISTINCT pelicula.id_director) > 1;
```

```python
%%sql
SELECT
    actor.nombre AS "Nombre"
FROM 
    actor
INNER JOIN 
    actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
INNER JOIN 
    pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
INNER JOIN
    director ON pelicula.id_director = director.id_director -- Corregido
WHERE 
    pelicula.id_director = 1
GROUP BY 
    actor.nombre
HAVING 
    COUNT(*) > 0;
```

**Solución:** Esta consulta recopila y muestra información sobre actores que no han ganado premios como mejor actor. Para cada actor listado, mostrando el nombre del actor, el nombre de una película en la que ha actuado, el nombre del director de esa película, el número total de películas diferentes dirigidas por ese director, y el número de premios que ese director ha ganado.

```python
%%sql
SELECT 
    actor.nombre AS "Nombre del actor",
    pelicula.nombre AS "Nombre de la película",
    director.nombre AS "Nombre del director",
    COUNT(DISTINCT pelicula.id_pelicula) AS "Número de películas del director",
    COUNT(DISTINCT premio_mejor_director.id_premio) AS "Número de premios del director"
FROM 
    actor AS actor
JOIN actor_participa_en_pelicula AS ap ON actor.id_actor = ap.id_actor
JOIN pelicula ON ap.id_pelicula = pelicula.id_pelicula
JOIN director ON pelicula.id_director = director.id_director
LEFT JOIN premio_mejor_director ON director.id_director = premio_mejor_director.id_director
WHERE actor.id_actor NOT IN (SELECT id_actor FROM premio_mejor_actor)
GROUP BY actor.nombre,
         pelicula.nombre,
         director.nombre;
```
